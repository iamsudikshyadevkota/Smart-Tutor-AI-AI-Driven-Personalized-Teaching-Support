```python
# Looping Through Pages from the website to scrape the data
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
import time

# Creating Lists
# Created four lists to store the data like bookname, authorname, ratings, pagenum from the url https://www.goodreads.com/list/show/10048.Non_fiction_Something_for_Everyone?page=1
bookname = []
authorname = []
ratings = []
pagenum= []

for i in range(1,24): # This loop iterates through the pages of the Goodreads list, from page 1 to page 23.

    # To know the time taken to scrape the data from one page, we have created an start_time attributes to store the start time before starting the scraping.
    start_time = time.time()

    # URL to scrape the data from
    URL = 'https://www.goodreads.com/list/show/10048.Non_fiction_Something_for_Everyone?page={}'.format(i)

    # Using the requests module, we are hitting the above web page, just like entering this url in any browser.
    page = requests.get(URL)

    # Now all the content from the webpage in the code formated is now stored in the pagecontent attribute.
    pagecontent = page.text

    # pageconent variable has unstructed which is very hard to get the required data out of it.
    # So using the Beautifulspoup module, we are organizing the HTML attributes and stroing them in the soup variable.
    soup = BeautifulSoup(pagecontent)

    # The above URL has the data in the table, so wea retrying the find the that table using its class name from the HTML attribute and stroed it in right_table variable.
    right_table=soup.find('table', class_='tableList js-dataTooltip')

    # Looping Through Table Rows:
    # As the table contains multiple data, we are looping through the right_table variable with all the data of HTML attribute type "tr", "tr" is the HTML table row attribute.
    # This loop iterates through each row (<tr>) of the table, extracting book information for each row.
    for row in right_table.findAll("tr"):

        # Extracting Data: Inside the loop, the code extracts book information like book name, author name, and rating using BeautifulSoup's find method with appropriate attributes.
        # From the table row, we are scraping the bookname, authroname, rating of book.
        # Scraping those data using the different HMTL attributes, which is for authorname attributename as itemprop and value for that attribute is name
        # Similary for authorname and rating, the attribute name is class and the value is authorname and minirating for authorname and book rating respectively.
        name = row.find('span', attrs = {'itemprop':'name'})
        author = row.find('a', attrs = {'class':'authorName'})
        rating = row.find('span', attrs = {'class':'minirating'})

        # As scraping is being done in a for loop for each page, we are storing the scraped data from each row in the array variable for all bookname, authorname and rating.
        # Appending Data to Lists:
        bookname.append(name.text)
        authorname.append(author.text)
        ratings.append(rating.text)

        #Adding the pagenumber from which we scraped the data to pagenum variable to understand whihc data is scraped from which page.
        pagenum.append(i)

    # Printing Progress:
    # Pring the time required to scrape the entire page using the start_time variable we created immediately after starting the for loop and before starting our web scraping process.
    print('Page {} completed in {} seconds'.format(i, time.time()-start_time))

# Creating a DataFrame: After scraping all the data, the script creates a Pandas DataFrame called books_df using the lists bookname, authorname, ratings, and pagenum.

# Creating the pandasdataframe object from the list generated after scraping all the data using the beautifulsoup function.
books_df = pd.DataFrame({'BookName':bookname, 'authorname':authorname, 'ratings': ratings, 'PageNo': pagenum})

# Displaying the First Rows:
books_df.head()
```

# Method 2 Scraping the data using API's from websites


This Python code is designed to scrape data from the Semantic Scholar API. It performs the following steps:

```python
# Scraping the different article data from the API provided by the sematic scholar

# Importing the required modules.
# These modules are imported to enable making HTTP requests and handling JSON data.
import requests
import json

# Defining the Search Topic:
# Creating the variable topic with the value as the keyword we need to search for in the semantic scholar articles.
# A search topic is defined as 'covid'. This keyword will be used to search for articles related to COVID-19.
topic = 'covid'

# Constructing the API URL:
# Storing the URL variable with dynamic query keywod you need to search for and the different queryparameters.
# To understand the queryparams, headers etc to hit the api in a correct way, check the following url.
# https://api.semanticscholar.org/api-docs/graph

# Meanwhile, through an API (Application Programming Interface), one can gain access to the data of an application or operating system.
# Therefore, APIs depend on the owner of the said dataset. The data can be either offered for free or be available at a cost.
# The owner can also limit the number of requests that a single user can make or the amount of data they can access.
url = "https://api.semanticscholar.org/graph/v1/paper/search?query="+ topic + "&offset=100&limit=3&fields=url,title,abstract,authors"

'''This line constructs the URL for the Semantic Scholar API with dynamic parameters:

query: The search keyword (in this case, 'covid').
offset: The starting index of the results (100 in this case).
limit: The number of results to retrieve (3 in this case).
fields: The fields to include in the response (URL, title, abstract, authors).
'''

# Creating Payload and Headers:
# Creating the empty payload and headers dictionary, as we are simply using the GET API here. Check the documentation.
# Empty dictionaries are created for the payload and headers. This example is using a simple GET request without any additional payload or headers.
payload={}
headers = {}

# Making a GET Request to the API:
# Hitting the GET api using the request module, with "GET" method as the API method, URL with the dynamic URL created in the previous step, headers and payload.
# The requests module is used to send a GET request to the Semantic Scholar API using the constructed URL, headers, and payload.
response = requests.request("GET", url, headers=headers, data=payload)

# Printing the Response Text:
# Just printing the response text from the above hit.
# The script prints the raw response text received from the API. This text initially contains JSON data as a string.
print(response.text)

# Parsing JSON Data:
#As the data we received is text, we need to convert them into the json using the json.loads fucntion.
#This conversion is possible as the text we received from GET method contains the formatted json data converted to string.
data = json.loads(response.text)['data']

# Creating a Pandas DataFrame:
#Created the pandas dataframe object using the json data we generated in previous step.
df = pd.DataFrame(data)
df.head()
```

The code starts by defining four variables: consumer_key, consumer_secret, access_token, and access_token_secret. These are authentication credentials required to access the Twitter API.

consumer_key and consumer_secret are used to identify the application (consumer) making the API requests.

access_token and access_token_secret are issued to the application after the user has authorized it to access their Twitter data. These tokens represent the user's access privileges.
